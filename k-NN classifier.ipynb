```python
import numpy as np
import matplotlib.pyplot as plt
import h5py

#Implement a function to compute the Euclidean distance between two vectors, and one to implement the k-NN algorithm by:
#   - Taking a sample
#   - Computing all the distances between the sample element and the elements of the training set
#   - sort the the training set based on the distances to the element (the use of functions like np.argsort is allowed)
#   - select the top k elements in terms of distance
#   - evaluate to which class the majority of these k elements belongs to (e.g., it is possible to use the function np.unique with the option return_counts=True and the function np.argmax)
```

```python
#Change the path to match the position of your file
#The Dataset can be loaded using thhe file option in Google Colab (the directory icon on the left)
Dataset1 = h5py.File('/content/drive/MyDrive/Colab Notebooks 2023 2024/Lab1_Ex_1_Synthtetic.hdf5')
Data = np.array(Dataset1.get('Dataset'))

Train_Set = Data[:200,:]
Test_Set = Data[200:,:]

#To be completed by the student
```

```python
accuracy_test = []
accuracy_train = []

#Function to calculate the euclidean distance

def euclidean_distance(x1, x2):
  #dist = np.linalg.norm(x1 - x2)
  dist = np.sqrt(np.sum(np.square(x1-x2)))
  return dist

#Prova

# for row in Data:
#   distance = euclidean_distance(Data[0], row)
#   print(distance)

#Function to calculate K-NN Algorithm
def KNN(sample, k):
  distances = []
  neighbors = []
  indexes = []

  for i in range(len(Train_Set)):
    distance = euclidean_distance(sample, Train_Set[i, :2]) #n_features = 2
    distances.append(distance) #vector containing all the distances

  #print(distances)
  indexes = np.argsort(distances) #neighbors containing the indexes of the distances
  #print(indexes)

  for i in range(k):
    neighbors.append(Train_Set[indexes[i], 2]) #Vector containing the classes of the k-nearest elements

  classes, counts = np.unique(neighbors, return_counts=True) #Evaluate the majority class; #classes is a vector with all unique classes; counts is a vector with the counter of all classes
  return classes[np.argmax(counts)] #return the index of the majority class; argmax(counts) is the index of the majority class

#Calculate correct prediction for train and test sets
for k in range(1, 200):
  prediction_test = 0
  prediction_train = 0
  for i in range(len(Test_Set)):
    if KNN(Test_Set[i, :2], k) == Test_Set[i, 2]:
      prediction_test += 1
    if KNN(Train_Set[i, :2], k) == Train_Set[i, 2]:
      prediction_train += 1
  accuracy_test.append(prediction_test/len(Test_Set))
  accuracy_train.append(prediction_train/len(Train_Set))

k=range(1, 200)
plt.plot(k, accuracy_test, "b", label="Test accuracy")
plt.plot(k, accuracy_train, "r", label="Train accuracy")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.show()





```

```python
#Part 1

Dataset2 = h5py.File('/content/drive/MyDrive/Colab Notebooks 2023 2024/Lab1_Ex_2_wine.hdf5')

Data = np.array(Dataset2.get('Dataset'))

Train_Set = Data[:200,:]
Test_Set = Data[200:,:]

#To be completed by the student
accuracy_train = []
accuracy_test = []

#Function to calculate the euclidean distance

def euclidean_distance(x1, x2):
  #dist = np.linalg.norm(x1 - x2)
  dist = np.sqrt(np.sum(np.square(x1-x2)))
  return dist

#Function to calculate K-NN Algorithm

def KNN(sample, k):
  distances = []
  neighbors = []
  indexes = []

  for i in range(len(Train_Set)):
    distance = euclidean_distance(sample, Train_Set[i, :11]) #n_feature = 11
    distances.append(distance) #vector containing all the distances

  indexes = np.argsort(distances) #neighbors containing the indexes of the distances

  for i in range(k):
    neighbors.append(Train_Set[indexes[i], 11]) #Vector containing the classes of the k-nearest elements

  classes, counts = np.unique(neighbors, return_counts = True) #Evaluate the majority class; #classes is a vector with all unique classes; counts is a vector with the counter of all classes

  return classes[np.argmax(counts)] #return the index of the majority class; argmax(counts) is the index of the majority class

for k in range(1, 200):
  prediction_test=0
  prediction_train=0
  for i in range(len(Test_Set)):
    if KNN(Test_Set[i,:11], k) == Test_Set[i,11]:
      prediction_test +=1
    if KNN(Train_Set[i,:11], k) == Train_Set[i,11]:
      prediction_train +=1
  accuracy_test.append(prediction_test/len(Test_Set))
  accuracy_train.append(prediction_train/len(Train_Set))

#plot

k = range(1, 200)
plt.plot(k, accuracy_test, "b", label='Test accuracy')
plt.plot(k, accuracy_train, "r", label='Train accuracy')
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.show()




```

**Exercise 3: Phoneme Dataset**

In this exercise the Phoneme dataset is examined https://catalog.ldc.upenn.edu/LDC93s1. Each line represents 256 samples gathered at a 16 kHz of different speech signals. The objective is to classify wether the sound emitted is a "sh", "iy", "dcl", "aa", "ao" phoneme.  

Again, a subset of the dataset containing 400 elements is provided. Create a training set and a test set of 200 samples each.


*   Classify the samples which compose the test set using the k-NN algorithm and evaluate the prediction accuracy for different values of k. Create a graph using the matplotlib library showing the evolution of the accuracy for different values of k over the test set.
*   Identifying overfitting and underfitting in the obtained results.

```python
Dataset3 = h5py.File('/content/drive/MyDrive/Colab Notebooks 2023 2024/Lab1_Ex_3.hdf5')

Data = np.array(Dataset3.get('Dataset'))

Train_Set = Data[:200,:]
Test_Set = Data[200:,:]

#To be completed by the student

accuracy_test = []
accuracy_train = []
prediction_test = 0
prediction_train = 0

def euclidean_distance(x1, x2):
  distance = np.sqrt(np.sum(np.square(x1 - x2)))
  return distance

def KNN(sample, k):
  distances = []
  neighbors = []
  indexes = []
  for i in range(len(Train_Set)):
    distances.append(euclidean_distance(sample, Train_Set[i, :256])) #n_features = 256
  indexes = np.argsort(distances)
  for i in range(k):
    neighbors.append(Train_Set[indexes[i],256])
  classes,counts = np.unique(neighbors, return_counts=True)
  return classes[np.argmax(counts)]

for k in range(1, 200):
  prediction_test = 0
  prediction_train = 0
  for i in range(len(Test_Set)):
    if KNN(Test_Set[i, :256], k) == Test_Set[i, 256]:
      prediction_test += 1
    if KNN(Train_Set[i, :256], k) == Train_Set[i, 256]:
      prediction_train += 1
  accuracy_test.append(prediction_test/len(Test_Set))
  accuracy_train.append(prediction_train/len(Train_Set))

k = range(1, 200)
plt.plot(k, accuracy_test, "b", label="Test accuracy")
plt.plot(k, accuracy_train, "r", label="Train accuracy")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.show()
```
